```python
from ultralytics import YOLO

# Choose the pre-trained YOLOv8 model you want to test
model = YOLO('my_model.pt')

# (Optional) If you want to work with the results in more detail
import numpy as np
import matplotlib.pyplot as plt

# Run validation on the COCO dataset
results = model.val(data='coco.yaml', save_json=True) 

# Print a summary of the metrics
print(results)

# Get individual metric values
metrics = results.box
mAP50 = metrics.map50  
mAP = metrics.map     
precision = metrics.precision.mean() 
recall = metrics.recall.mean() 

print(f'mAP50: {mAP50:.3f}')
print(f'mAP: {mAP:.3f}')
print(f'Precision: {precision:.3f}')
print(f'Recall: {recall:.3f}')
```

```python
from ultralytics import YOLO
import numpy as np

# Choose the pre-trained YOLOv8 model you want to test
model = YOLO('my_model.pt')

# Run validation on the COCO dataset (replace with your dataset path if different)
results = model.val(data='coco.yaml', save_json=True)

# Print a summary of the metrics
print(results)

# Get individual metric values
metrics = results.box
mAP50 = metrics.map50
mAP = metrics.map

# Calculate mean precision and recall from the lists
precision = np.mean(metrics.p)  # Average precision over all classes
recall = np.mean(metrics.r)     # Average recall over all classes

print(f'mAP50: {mAP50:.3f}')
print(f'mAP: {mAP:.3f}')
print(f'Precision: {precision:.3f}')
print(f'Recall: {recall:.3f}')
```
